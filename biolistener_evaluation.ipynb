```python
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [10, 5]
matplotlib.rcParams['figure.dpi'] = 200

from biolistener_evaluation import CMRRAnalyzer, IRNAnalyzer, FrequencyResponseAnalyzer, SNRAnalyzer
```

## ADC AD7771 (Analog Devices)
#### Input Referred Noise (IRN)

```python
irn_analyzer = IRNAnalyzer('./data/irn/streamer_2024_12_25_21_55_32_ad7771_test_2.csv')

irn_analyzer.display_results()
```

```python
irn_analyzer.plot()
```

Short summary: IRN (STD ±3σ 99.73%) is below 2uV for all channels, this is a good result.

#### Frequency Response

```python
frequency_response_analyzer = FrequencyResponseAnalyzer('./data/frequency_response_ad7771')
frequency_response_analyzer.plot()
```

Short summary: Frequency response is within the range required by regulations for EEG and ECG (thus for all other common biomedical applications as well).


#### Signal to Noise Ratio (SNR)

```python
snr_analyzer = SNRAnalyzer('./data/frequency_response_ad7771')
snr_analyzer.plot()
```

Short summary: SNR is great, comparable to costly EEG devices.


#### Common Mode Rejection Ratio (CMRR) - Balance test

```python
cmrr_balance_analyzer = CMRRAnalyzer('./data/cmrr/cmrr_balance_ad7771')
cmrr_balance_analyzer.plot()
```

Short summary: CMRR balance test results are great, above measurable with common equipment. Professional equipment is needed to measure this parameter.


#### Common Mode Rejection Ratio (CMRR) - Imbalance test

```python
cmrr_imbalance_analyzer = CMRRAnalyzer('./data/cmrr/cmrr_imbalance_ad7771')
cmrr_imbalance_analyzer.plot()
```

Short summary: CMRR is great for all common biomedical applications.


#### Inputs impedance
More than 10MOhm.
